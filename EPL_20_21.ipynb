```python
#import the liabraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load Dataset

epl_df=pd.read_csv(r'C:\Users\29ske\Desktop\Analytics With Python Project\EPL.csv')
epl_df.head()
```

```python
epl_df.info()
```

```python
#Data Description, works only on numeric data

epl_df.describe()
```

```python
#Check is there any null value 

epl_df.isna().sum()
```

```python
#Creat two new columns 

epl_df['MinsPerMatch']=(epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch']=(epl_df['Goals']/epl_df['Matches']).astype(float)
epl_df.head()
```

```python
#Total Goals 

TotalGoals = epl_df['Goals'].sum()
TotalGoals
```

```python
#Penalty Goals 

ToatalPenaltyGoals = epl_df['Penalty_Goals'].sum()
ToatalPenaltyGoals
```

```python
#Penalty Attempts 

TotalPeNaltyAttempts = epl_df['Penalty_Attempted'].sum()
TotalPeNaltyAttempts

#125 penalty attemps were there from which 102 got converted 
```

```python
#Piechart for penalties missed vs penalties scored

plt.figure(figsize=(13,6))
PlNotScored = TotalPeNaltyAttempts - ToatalPenaltyGoals
data = [PlNotScored,ToatalPenaltyGoals]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.legend()
plt.show()
```

```python
#Unique Positions 
epl_df['Position'].unique()
```

```python
#Total number of FW Players

epl_df[epl_df['Position']=='FW']
```

```python
#Players from different Nations 

np.size(epl_df['Nationality'].unique())
```

```python
#Most players from which countries 

MostPlayerNation = epl_df.groupby('Nationality').size().sort_values(ascending =False)
MostPlayerNation.head(10).plot(kind = 'bar', figsize=(12,6), color = sns.color_palette('magma'))
```

```python
#Clubs with maximum player in their squad 
epl_df['Club'].value_counts().nlargest(5).plot(kind='bar',color=sns.color_palette('cubehelix'))
```

```python
#Clubs with least players in their squad

epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar',color=sns.color_palette('viridis'))
```

```python
#Players based on age bracket 

Under20 = epl_df[epl_df['Age'] <= 20]
Age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
Age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
AgeAbove30 = epl_df[epl_df['Age'] > 30]
```

```python
#Pie chart of the players according to age brackets 

x = np.array([Under20['Name'].count(),Age20_25['Name'].count(),Age25_30['Name'].count(),AgeAbove30['Name'].count()])
mylabels = ['<=20','>20 & <=25','>25 & <=30', '>30']
plt.title('Total Players With Age brackets', fontsize = 20)
color = sns.color_palette("colorblind")
plt.pie(x,labels= mylabels,colors = color, autopct ='%0.1f%%')
#plt.legend(loc='upper right')

plt.show()
```

```python
#Total under 20 players in each club

PlayersUnder20 = epl_df[epl_df['Age']<20]
PlayersUnder20['Club'].value_counts().plot(kind = 'bar', color=sns.color_palette('viridis'))
```

```python
#Players under 20 plays in Manchester United

PlayersUnder20[PlayersUnder20['Club']=='Manchester United']
```

```python
#Under 20players in chelsea
PlayersUnder20[PlayersUnder20['Club']=='Chelsea']
```

```python
#Average Age of players in each club 

plt.figure(figsize=(12,6))
sns.boxplot(x = 'Club', y='Age', data= epl_df)
plt.xticks(rotation = 90)
```

```python
#number of players in each club according to age in ascending order 
NumPlayer = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum())/NumPlayer
data.sort_values(ascending = False)
```

```python
# Total Assist from each club

AssistByClubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style = 'whitegrid',color_codes= True)
ax = sns.barplot(x='Club', y='Assists', data=AssistByClubs.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(16,8)
plt.title('Plot of Cubs vs Total Assists', fontsize = 20)
```

```python
#Top 10 Assists by players

Top10Assists = epl_df[['Name','Club', 'Assists', 'Matches']].nlargest(n=10, columns = 'Assists')
Top10Assists
```

```python
#Total Goals from each club 

GoalsbyClubs = pd.DataFrame(epl_df.groupby('Club', as_index= False)['Goals'].sum())
sns.set_theme(style = 'whitegrid',color_codes= True)
ax = sns.barplot(x='Club', y='Goals', data=GoalsbyClubs.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel('Club',fontsize=30)
ax.set_xlabel('Goals',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(12,8)
plt.title('Plot of Cubs vs Total Goals', fontsize = 20)
```

```python
# Most Goals by players 

Top10Goals = epl_df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=10, columns = 'Goals')
Top10Goals
```

```python
# Goals Per match

Top10GoalsPerMatch =epl_df[['Name', 'GoalsPerMatch', 'Matches','Goals']].nlargest(n=10, columns = 'GoalsPerMatch')
Top10GoalsPerMatch
```

```python
#pie chart - Goals with assists and without assists 

plt.figure(figsize=(14,7))
assists = epl_df['Assists'].sum()
data = [TotalGoals - assists, assists]
labels = ['Goals without assists ', 'Goals with assists']
color= sns.color_palette('Set1')
plt.pie(data, labels = labels, colors= color,autopct = '%.0f%%')
plt.show()
```

```python
#Top 10 players with most yellow cards 

epl_yellow = epl_df.sort_values(by= 'Yellow_Cards', ascending = False)[:10]
plt.figure(figsize=(12,6))
plt.title('Players with the most yellow cards')
c=sns.barplot(x=epl_yellow['Name'], y=epl_yellow['Yellow_Cards'], label = 'Players', color= 'yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(),rotation = 45)
c
```
